SCRAPING THE DATA FROM THE WEBPAGE

```python
data = codecs.open('List_of_postal_codes_of_Canada:_M')
html = data.read()
soup = BeautifulSoup(html, 'html.parser')
tor = (pd.read_html(str(soup.table))[0])
#Code referred from http://beautiful-soup-4.readthedocs.io/en/latest/
```

ASSIGNING COLUMN NAMES APPROPRIATELY

```python
#tor is the name of my dataframe
tor.drop(axis=0, index=0, inplace=True)
col_names = ['Postcode', 'Borough', 'Neighborhood'] 
tor.columns = col_names
tor
```

REMOVING NA VALUES

```python
tor = tor[tor['Borough'] != 'Not assigned'].reset_index(drop=True) 
tor
```

```python
na = tor[tor['Neighborhood']=="Not assigned"]
for index in range(len(na)):
    row = na.iloc[index]
    row['Neighborhood']=row['Borough']

tor
```

MERGING THE ROWS WITH SIMILAR POST CODES

```python
aggregations = {'Neighborhood':lambda x:', '.join(x)} #dictionary comprehension
group = tor.groupby('Postcode').agg(aggregations).reset_index()
x = tor.drop('Neighborhood',axis=1)
merged_data = pd.merge(x, group, on='Postcode', how='left')
tor1 = merged_data.drop_duplicates()
tor1.to_csv("Toronto.csv", index=False)
tor1
```
